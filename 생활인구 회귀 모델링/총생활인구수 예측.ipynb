```python
import pandas as pd
data=pd.read_csv("생활인구 회귀.csv",encoding='CP949')
del data['Unnamed: 0']
csv=pd.read_csv("randomforest_prediction.csv",encoding='CP949')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
```

```python
data.columns
```

```python
np.random.seed(1003)
```

```python
y=data['총생활인구수']
x=data[['2022년06월_총인구수','편의점','카페','음식점','토지 매매가']]

from sklearn.model_selection import train_test_split
X_train,X_validation,y_train,y_validation=train_test_split(
    x,y,test_size=0.2,random_state=1003)
```

```python
regressor=RandomForestRegressor(n_estimators=300,max_features=3)
regressor.fit(X_train, y_train)
```

```python
x=csv[['2022년06월_총인구수','편의점','카페','음식점','토지 매매가']]
pred=regressor.predict(x)
```

```python
pd.options.display.float_format = '{:.5f}'.format
pd.DataFrame({'총 생활인구수 예측값':pred})
```

```python
csv['총 생활인구수 예측값']=pd.DataFrame({'총 생활인구수 예측값':pred})
csv
```

```python
csv.to_csv('총 생활인구수 예측.csv',encoding='CP949',index=False)
```
